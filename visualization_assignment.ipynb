```python
import matplotlib.pyplot as plt
import numpy as np
import os
if not os.path.exists('imgs'):
    os.makedirs('imgs')

plt.style.use('seaborn-v0_8-muted')
```

# Gráfico de Líneas: Evolución de Ventas Mensuales
### Este gráfico muestra la tendencia de ventas proyectadas a lo largo de un año. Se eligió un gráfico de líneas porque es la herramienta ideal para visualizar series temporales y cambios en variables continuas, permitiendo identificar rápidamente picos y caídas estacionales.

```python
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas_a = [120, 135, 150, 145, 160, 180, 210, 195, 170, 160, 190, 230]
ventas_b = [90, 105, 110, 125, 140, 150, 160, 175, 180, 190, 210, 240]

plt.figure(figsize=(10, 6))
plt.plot(meses, ventas_a, label='Producto A', color='#2E86C1', marker='o', linestyle='-', linewidth=2)
plt.plot(meses, ventas_b, label='Producto B', color='#E67E22', marker='s', linestyle='--', linewidth=2)

plt.title('Tendencia de Ventas Mensuales (2023)', fontsize=14)
plt.xlabel('Meses del Año')
plt.ylabel('Ventas (en miles de USD)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Gráfico de Barras: Rendimiento por Departamento
### Aquí comparamos el presupuesto asignado frente al gasto real de cinco departamentos de una empresa. El gráfico de barras es excelente para comparar magnitudes entre categorías discretas, facilitando la identificación de desviaciones presupuestarias.

```python
depts = ['IT', 'RRHH', 'Ventas', 'Marketing', 'I+D']
presupuesto = [50, 25, 45, 60, 40]
gasto_real = [48, 28, 42, 65, 38]

x = np.arange(len(depts))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width/2, presupuesto, width, label='Presupuesto', color='#1ABC9C')
ax.bar(x + width/2, gasto_real, width, label='Gasto Real', color='#C0392B')

ax.set_title('Presupuesto vs. Gasto Real por Departamento', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(depts)
ax.set_ylabel('Monto (Millones $)')
ax.legend()

plt.savefig('imgs/bar_chart.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Histograma: Distribución de Calificaciones
### Este gráfico representa la distribución de frecuencias de las calificaciones de 500 estudiantes en un examen final. Se utiliza un histograma para entender la forma de los datos y localizar dónde se concentra la mayor parte de la población.

```python
np.random.seed(42)
calificaciones = np.random.normal(75, 10, 500)

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(calificaciones, bins=20, color='#9B59B6', edgecolor='white', alpha=0.7)

plt.title('Distribución de Calificaciones del Examen Final', fontsize=14)
plt.xlabel('Puntaje (0 - 100)')
plt.ylabel('Número de Estudiantes')
plt.axvline(calificaciones.mean(), color='red', linestyle='dashed', linewidth=2, label=f'Promedio: {calificaciones.mean():.1f}')
plt.legend()

plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Scatter Plot: Relación Publicidad vs. Ventas
### Se analiza la correlación entre la inversión en publicidad y el retorno en ventas. El gráfico de dispersión (Scatter Plot) es el estándar para investigar la relación entre dos variables numéricas

```python
inversion = np.random.rand(50) * 100
ventas = (inversion * 1.5) + np.random.normal(0, 15, 50)

plt.figure(figsize=(10, 6))
plt.scatter(inversion, ventas, c=ventas, cmap='viridis', s=100, alpha=0.7, edgecolors='black')

plt.title('Correlación: Inversión en Publicidad vs. Ventas', fontsize=14)
plt.xlabel('Inversión publicitaria (USD)')
plt.ylabel('Ventas generadas (USD)')
plt.colorbar(label='Escala de Ventas')
plt.grid(True, linestyle=':', alpha=0.6)

plt.savefig('imgs/scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Gráfico Avanzado: Comparativa de Funciones Trigonométricas
### En esta visualización avanzada utilizamos subplots para mostrar la función Seno y Coseno de forma paralela. Se eligió este formato para permitir una comparación técnica directa entre dos funciones


```python
x_val = np.linspace(0, 10, 100)
y_sin = np.sin(x_val)
y_cos = np.cos(x_val)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Subplot 1: Seno (Nota la 'r' antes de las comillas en el label)
ax1.plot(x_val, y_sin, color='#3498DB', label=r'$\sin(x)$')
ax1.set_title('Función Seno')
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Amplitud')
ax1.grid(True)
ax1.legend()

# Subplot 2: Coseno (Nota la 'r' antes de las comillas en el label)
ax2.plot(x_val, y_cos, color='#E74C3C', label=r'$\cos(x)$')
ax2.set_title('Función Coseno')
ax2.set_xlabel('Tiempo (s)')
ax2.grid(True)
ax2.legend()

plt.suptitle('Análisis de Oscilaciones Trigonométricas', fontsize=16)
plt.tight_layout()

plt.savefig('imgs/custom_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
